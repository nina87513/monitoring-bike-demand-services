```python
bikeshare_preprocessing = pd.DataFrame()
status_data = dict()
for i in status.station_id.unique():
    status_i = status.loc[status['station_id']==i, :]
    
    status_i.loc[:, 'month'] = pd.to_datetime(status_i.time, format='%Y-%m-%d %H:%M:%S').dt.month
    status_i.loc[:, 'dayofweek'] = pd.to_datetime(status_i.time, format='%Y-%m-%d %H:%M:%S').dt.dayofweek
    status_i.loc[:, 'hour'] = pd.to_datetime(status_i.time, format='%Y-%m-%d %H:%M:%S').dt.hour
    status_i.loc[:, 'time'] = pd.to_datetime(status_i['time'].str[:10]).dt.strftime('%Y/%m/%d')
    status_i = status_i.groupby(['time', 'hour']).mean()
    status_i.reset_index(inplace=True, level=['time', 'hour'])
    status_i['dock_count'] = station.loc[station.id == i, 'dock_count'].values[0]

    status_i = pd.merge(status_i, weather[weather['zip_code'] == status_i.city_zip.unique()[0]], left_on=['time', 'city_zip'], right_on=['date', 'zip_code'])
    status_i = status_i.drop(['date', 'zip_code'], axis=1)
    
    status_data[i] = status_i
    bikeshare_preprocessing = bikeshare_preprocessing.append(status_i)
    print(f"station {i} done, shape: {status_i.shape}")
    print(status_i.loc[0, ['time', 'hour']])
```
